```python
# importamos las librerías que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd

# Visualización
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?

```python
# fertilidad por país
sns.histplot(x = "fertility_rate", data = df, bins = 10, kde = True, color = "violet");
```

```python
sns.countplot(x = "fertility_rate", data = df)
```

```python
sns.boxplot(x = "fertility_rate", 
            data = df, 
            width = 0.5, 
            color = "turquoise");
```

```python
sns.violinplot(x = "fertility_rate", 
            data = df, 
            width = 0.5, 
            color = "turquoise");
```

¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? Proporciona una breve descripción de los patrones que observes.

```python
df.columns
```

```python
sns.scatterplot(x = "infant_mortality", 
                y = "life_expectancy", 
                data = df,
                size = 200, 
                marker= "v", 
                color = "orange")
                

# cambiamos los nombres de los ejes
plt.xlabel("infant_mortality")
plt.ylabel("life_expectancy");
```

```python
sns.regplot(x = "infant_mortality", 
            y = "life_expectancy", 
            data = df, 
            marker = "d",
            line_kws = {"color": "black", "linewidth": 1},
            scatter_kws = {"color": "teal", "s": 1} 
            )

# cambiamos los nombres de los ejes como hemos estado haciendo hasta ahora
plt.xlabel("infant_mortality", fontsize = 8)
plt.ylabel("life_expectancy", fontsize = 8)

# ponemos título a la gráfica
plt.title("Relación entre tasa de mortalidad infantil y la esperanza de vida", fontsize = 8);
```
